# ZOMATO DATA ANALYSIS PROJECT

## Overview

The number of restaurant in Bengaluru are increasing day by day. Currently which stands at approximately 12,000 restaurants. With such an high number of restaurants this industry hasn't been saturated yet and new restaurants are opening every day. However it has become difficult for them to compete with already established restaurants. The key issues that continue to pose a challenge to them include high real estate costs, rising food costs, shortage of quality manpower, fragmented supply chain and over-licensing. This Zomato data aims at analysing demography of the location. Most importantly it will help new restaurants in deciding their theme, menus, cuisine, cost etc for a particular location.

### GOAL

Analyze the zomato dataset and try to figure out what things a restaurant owner should keep in mind before opeining their new restaurant in Bengaluru 

Find the theme, menus, cuisine, cost etc for a particular location.

zomato.csv

```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the csv file
df = pd.read_csv('zomato.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
# dropping columns which might not be very useful in analyzing data
df = df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'], axis = 1)
```

```python
df.head()
```

```python
df.info()
```

```python
# dropping the duplicates
df= df.drop_duplicates()
df.shape
```

```python
# cleaning rate column
df['rate'].unique()
```

```python
def newrate(value):
    if(value == 'NEW' or value == '-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
```

```python
df['rate'] = df['rate'].apply(newrate)
```

```python
df['rate'].unique()
```

```python
df['rate'].isnull().sum()
```

```python
df['rate'] = df['rate'].fillna(df['rate'].mean())
```

```python
df['rate'].isnull().sum()
```

```python
# dropping the duplicates
df = df.dropna()
```

```python
df.head()
```

```python
df.info()
```

```python
# renaming the columns
df = df.rename(columns = {'approx_cost(for two people)': 'cost/2plate','listed_in(type)':'Type', 'listed_in(city)':'city'})
```

```python
df.head()
```

```python
# dropping city column
df = df.drop('city', axis = 1)
```

```python
df.head()
```

```python
df['location'].unique()
```

```python
# cleaning cost/plate column
df['cost/2plate'].unique()
```

```python
def removecomma(x):
    x = str(x)
    if ',' in x:
        x = x.replace(',','')
        return float(x)
    else:
        return float(x)
    
```

```python
df['cost/2plate'] = df['cost/2plate'].apply(removecomma)
```

```python
df['cost/2plate'].unique()
```

```python
df.head()
```

```python
# cleaning and handling rest type column
df['rest_type'].unique()
```

```python
df.head(10)
```

```python
rest_types = df['rest_type'].value_counts(ascending= False)
```

```python
rest_types
```

```python
# clustering restaurant type less than 1000 as others
rest_types_lessthan1000 = rest_types[rest_types < 1000]
```

```python
rest_types_lessthan1000 
```

```python
def restaurant_type(x):
    if x in rest_types_lessthan1000:
        return 'others'
    else:
        return x
```

```python
df
```

```python
df['rest_type'] = df['rest_type'].apply(restaurant_type)
```

```python
df['rest_type'].value_counts()
```

```python
# cleaning and handling location column
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending= False)
location_lessthan300 = location[location<300]
```

```python
def locate(x):
    if x in location_lessthan300:
        return 'others'
    else:
        return x
```

```python
df['location'] = df['location'].apply(locate)
```

```python
df['location'].value_counts()
```

```python
# handling and cleaning cuisines column
df['cuisines'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending= False)
```

```python
cuisines_lessthan100 = cuisines[cuisines<100]
```

```python
def cuisine_clean(x):
    if x in cuisines_lessthan100:
        return 'others'
    else:
        return x
```

```python
df['cuisines'] = df['cuisines'].apply(cuisine_clean)
```

```python
df
```

```python
df['cuisines'].value_counts()
```

```python
# analyzing type column
df['Type'].value_counts()
```

## CONCLUSION

Through the above data that we have analyzed we can say that if someone is trying to opeining a new restaurant in Bengaluru then they should keep the following things in mind

- BTM is a very competitive area as there are so many restaurants already opened there which might make it a very saturated place to open a new restaurant

- Old airport road, Shanti Nagar, St. Marks road might be better place to open restaurants as they have less than 500 restaurants opened and it might not be very competitive

- Most of the restaurants do not provide book table facility

- Average rating of restaurant which provide online orders and which do not provide online orders are same

- Average rating of restaurant which provide book table facility is higher than restaurants which do not provide book table facility

- Most of the restaurants in Old airport road, Shanti Nagar, St. Marks road and Shivaji nagar do not provide online order facility

- In almost every location very few restaurants provide book table facility

- Pubs and bars, drinks and nightlife got the highest ratings out of every type of restaurant

- Shivaji, Shanti Nagar, St. Marks road and even BTM and Banashankari has very few Pubs and bars, drinks and nightlife restaurants making it good location to open those kind of restaurants there
